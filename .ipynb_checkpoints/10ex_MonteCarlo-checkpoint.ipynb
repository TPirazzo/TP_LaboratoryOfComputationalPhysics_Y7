1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
# First point 
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
%precision 4

# Constants
tau = 3.052 * 60    # Convert half-life to seconds 
decay_constant = np.log(2) / tau 

# Initial conditions 
Tl = 1000 
Pb = 0 
t_steps = 1500

# Arrays to store the number of atoms at each time step 
Tl_t = np.zeros(t_steps + 1) 
Pb_t = np.zeros(t_steps + 1) 
Tl_t[0] = Tl  #array to store the number of remainig Tallium atoms
Pb_t[0] = Pb  #array to store the number of remainig Lead atoms

# Simulation 
for t in range(1, t_steps + 1):
    decayed_atoms = 0 
    for atom in range(int(Tl_t[t-1])): 
        if np.random.uniform() < 1 - 2**(-1/tau):
            decayed_atoms += 1 
    Tl_t[t] = Tl_t[t-1] - decayed_atoms 
    Pb_t[t] = Pb_t[t-1] + decayed_atoms 
            
# Time array
time = np.arange(0, t_steps + 1) 

# Plotting 
plt.plot(time, Tl_t, label='Thallium (${\\rm Tl}^{208}$)') 
plt.plot(time, Pb_t, label='Lead (${\\rm Pb}^{208}$)') 
plt.xlabel('Time (seconds)') 
plt.ylabel('Number of atoms') 
plt.legend() 
plt.title('Decay of ${\\rm Tl}^{208}$ to ${\\rm Pb}^{208}$')
```

```python
# Second point

# Define the inverse-cumulative-pdf
def icdf(s,T = tau):
    return -T*np.log2(1-s)
    
def cdf(s,T = tau):
    return 1-2**(-s/T)

# show the cumulative probability function from which we draw the random numbers in order to calculate the decay times of the Tallium atoms
x = np.linspace(0,1500,10000)
y = np.linspace(0,1,100)
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(x, cdf(x))
plt.ylabel('1: Generate a (0,1) uniform PRNG')
plt.xlabel('2: Find the inverse CDF')
plt.title('Inverse transform method');

# plot the number of not decayed Tl atoms 
plt.subplot(122)
z = np.random.uniform(0,1,Tl)
sorted_decay_times = np.sort(icdf(z))
atoms_not_decayed = Tl - np.arange(1, Tl+1)
plt.plot(sorted_decay_times, atoms_not_decayed, label='Thallium (${\\rm Tl}^{208}$)')
plt.xlabel('Time (seconds)') 
plt.ylabel('Number of atoms') 
plt.legend() 
plt.title('Number of not decayed ${\\rm Tl}^{208}$ atoms as function of time')
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
# Constants
Z = 79 
e = 1.602e-19 # Charge of an electron in C 
epsilon_0 = 8.854e-12 # Permittivity of free space in F/m 
E = 7.7e6 * 1.602e-13 # Energy in Joules (convert MeV to J) 
a0 = 5.29177e-11 # Bohr radius in meters
N = int(1e6) # Number of alpha particles to simulate

# Define the inverse-cumulative-pdf
def icdf_R(s,sigma = a0/100):
    return np.sqrt(-2*(sigma**2)*np.log(1-s))

# Calculate the critical impact parameter 
b_crit = (Z * e**2) / (2 * np.pi * epsilon_0 * E) 
print("Critical value of the impact parameter b_crit: ", b_crit, "m", '\n')

# Draw the random numbers
z = np.random.uniform(0,1,N)
sorted_impact_parameter = np.sort(icdf_R(z))
plt.hist(sorted_impact_parameter, histtype='step', bins=100, density=True, linewidth=2)

# Plot the distribution of b 
#plt.plot(sorted_impact_parameter)
plt.xlabel('impact parameter b [m]') 
plt.ylabel('Number of alpha particles') 
plt.title('Distribution of the impact parameter')

# the alpha particles that "bounce back" are those wich b < (Ze^2)/(2*pi*epsilon_0*E) = b_crit
fraction_bounce_back = np.sum(sorted_impact_parameter < b_crit) / N # Fraction of particles that bounce back 
print("Fraction of particles that bounce back: ", fraction_bounce_back, '\n')
num_particles_bounce_back = fraction_bounce_back * N # Number of particles that bounce back  
print("Number of particles that bounce back: ", num_particles_bounce_back, '\n')
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
# Compute the integral of f(x) between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
def f(x):
    return (np.sin(1/(x*(2-x))))**2

A = 2*1 #box

x=np.linspace(0.001,1.999,1000)
plt.plot(x,f(x),'r-')

# Monte Carlo integration
N=100000
count=0
for i in range(N):
    x=np.random.uniform(0,2,1)
    y=np.random.random()
    if y<f(x): count+=1
I=A*count/N
hit_miss_err = A * np.sqrt((count / N) * (1 - count / N) / N) # Binomial variance err = radq(N*p*(p-1)) p->sucsess prob
print("The integral value estimation with the hit/miss method is: ",I,"+/-","{:.5f}".format(hit_miss_err))

# Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one
def mean_f(x):
    return (1/len(x))*sum(f(x))

r = np.random.uniform(0,2,N) 
I_m = 2*mean_f(r)
mean_method_err = (2/np.sqrt(N))*np.var(f(r)) # Formula in the lesson's notebook
print("The integral value estimation with the mean value method is: ",I_m,"+/-","{:.5f}".format(mean_method_err))
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
# Draw random points from a square centered in (0,0) and l=2
N = 1000
x_r = np.random.uniform(-1,1,N)
y_r = np.random.uniform(-1,1,N)
r = x_r**2 + y_r**2
points_inside = np.sum(r <= 1)

# Using the hint/miss model
A = 2*2 # x,y -> [-1,1],[-1,1]
I = A*points_inside/N
print("The area of the circle is: ",I)
 
# In order to generalize the process to a 10D sphere we do the same thing with 10 variables
D = 10
A_10D = 2**D
points = np.random.uniform(-1, 1, (N, D)) 
distances = np.sum(points**2, axis=1) 
points_inside_sphere = np.sum(distances <= 1) 
volume_estimate = A * points_inside_sphere / N 
print(f"Stima del volume della sfera di raggio unitario in {D} dimensioni: {volume_estimate}") 
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
N = 10000

# Draw the weight from the given distribution with the icdf method
u = np.random.uniform(0,1,N)
x_r = (u/2)**2 # icdf(x)=(x/2)^2 ; cdf(x)=2*sqrt(x) 
plt.plot(np.sort(x_r))

# Define the function to integrate
def f(x): 
    return 1 / (np.exp(x) + 1) 

# Compute the mean value of the integral 
integral_estimate = 1.0/N*np.sum(f(x_r))*2

# Error estimation
error_estimate = np.std(f(x_r)) / np.sqrt(N) 
print(f"Stima dell'integrale: {integral_estimate}") 
print(f"Errore stimato: {error_estimate}") 
```

```python
# Numero di campioni 
N = 100000

# Genera campioni secondo la distribuzione w(x) = 1/sqrt(x) 
x_samples = np.random.uniform(0, 1, N)**2 

# Definisce la funzione da integrare f(x) 
def f(x): return 1 / (np.exp(x) + 1) 

# Calcola i pesi corretti 
weights = 1 / np.sqrt(x_samples) 

# Calcola la stima dell'integrale utilizzando il campionamento per importanza 
integral_estimate = np.mean(f(x_samples) * weights)

# Calcola l'errore 
error_estimate = np.std(f(x_samples) * weights) / np.sqrt(N) 
print(f"Stima dell'integrale: {integral_estimate:.4f}") 
print(f"Errore stimato: {error_estimate:.4f}")
```
